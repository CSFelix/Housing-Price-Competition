<b>Mutual Information (MI)</b>

````python
from sklearn.feature_selection import mutual_info_regression
````

````python
# Function to Plot MI Scores
def plot_mi_scores(scores):
    """
    Plots Mutual Information Scores in Ascending Order
    """
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")
    plt.show()
````

````python
# Getting the discrete (integer) and continuous (float) features
discrete_features = [
    feature for feature in good_labels_X_train.columns
    if good_labels_X_train[feature].dtype == 'int64'
]

continuous_features = [
    feature for feature in good_labels_X_train.columns
    if good_labels_X_train[feature].dtype == 'float64'
]

# Creating copy datasets for discrete and continuous features
discrete_features_list = good_labels_X_train[discrete_features].dtypes == 'int64'
continuous_features_list = good_labels_X_train[continuous_features].dtypes == 'float64'
````

````python
# Calculating Features MI Scores
mi_scores_discrete_features = mutual_info_regression(good_labels_X_train_temp[discrete_features], y_train, discrete_features=discrete_features_list, random_state=SEED)
mi_scores_discrete_features = pd.Series(mi_scores_discrete_features, name='MI Scores - Discrete Features', index=discrete_features)
mi_scores_discrete_features = mi_scores_discrete_features.sort_values(ascending=False)

mi_scores_continuous_features = mutual_info_regression(good_labels_X_train_temp[continuous_features], y_train, discrete_features=continuous_features_list, random_state=SEED)
mi_scores_continuous_features = pd.Series(mi_scores_continuous_features, name='MI Scores - Continuous Features', index=continuous_features)
mi_scores_continuous_features = mi_scores_continuous_features.sort_values(ascending=False)

# Plotting the Results
plt.figure(dpi=100, figsize=(15, 20))
plot_mi_scores(mi_scores_discrete_features)

plt.figure(dpi=100, figsize=(10, 10))
plot_mi_scores(mi_scores_continuous_features)
````

<b>K-Means Clustering</b>

````python
from sklearn.cluster import KMeans
````

````python
# First, let's find out what's the perfect number of clusters applying
# the Elbow and WCSS Methods
wcss = []

for i in range(1,7):
    kmeans = KMeans(
        n_clusters=i
        , init='k-means++'
        , max_iter=300
        , n_init=10
        , random_state=SEED
    )
    kmeans.fit(good_labels_X_train_temp)
    wcss.append(kmeans.inertia_)
    
# Plotting the results
plt.plot(range(1,7), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
````

We have elbows with **2, 3, 4 and 5 clusters** Let's use **4 clusters** in this project!!

````python
kmeans = KMeans(
    n_clusters=4
    , init='k-means++'
    , max_iter=300
    , n_init=10
    , random_state=SEED
)

good_labels_X_train_temp.insert(loc=0, column='Cluster', value=kmeans.fit_predict(good_labels_X_train_temp))
good_labels_X_train_temp.head()
````

````python
# Pipelines
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
from sklearn.compose import ColumnTransformer

# Numerical Features Transformer
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant'))
#    , ('scaler', StandardScaler())
#   ,  ('scaler', RobustScaler())
   ,  ('scaler', MinMaxScaler())
#   ,  ('scaler', Normalizer())
])

# Bundling Preprocessors
preprocessor = ColumnTransformer(
    transformers=[('numerical', numerical_transformer, good_labels_X_train.columns)]
)
````

````python
# Dropping Features
features_to_drop = [
    'LotConfig', 'MasVnrType' , 'MasVnrArea'
    , 'MoSold'   , 'YrSold', 'SaleType'
    , 'SaleCondition', 'Electrical'
    , 'GarageQual', 'HeatingQC', 'RoofMatl'
    , 'ExterCond', 'HouseStyle', 'LotShape'
    , 'GarageCond', 'CentralAir', 'HalfBath'
    , 'BedroomAbvGr', 'PavedDrive', 'BsmtFullBath'
    , 'EnclosedPorch', 'BsmtCond', 'Alley'
    , 'BldgType', 'Condition1', 'Fence'
    , 'RoofStyle', 'Functional', 'ScreenPorch'
    , 'KitchenAbvGr', 'Heating', 'BsmtFinType2'
    , 'BsmtHalfBath', 'LandContour', 'MiscFeature'
    , 'LowQualFinSF', 'MiscVal', '3SsnPorch'
    , 'Street', 'Condition2', 'LandSlope'
    , 'PoolArea', 'PoolQC', 'Utilities'
    , 'MSZoning', 'BsmtFinSF1', 'BsmtFinSF2'
]

test_dataset.drop(features_to_drop, axis=1, inplace=True)
````

````python
# Converting Categorical Features into Lower Case
categorical_features = [
    col for col in test_dataset.columns
    if test_dataset[col].dtypes == 'object'
]

for column in categorical_features:
    test_dataset[column] = test_dataset[column].str.lower()
    
test_dataset[categorical_features].head()
````

````python
# Encoding Categorical Features
categorical_features_unique_values = { }

for column in categorical_features:
    categorical_features_unique_values[column] = list(enumerate(test_dataset[column].unique()))

for feature in categorical_features:
    for index in range(len(categorical_features_unique_values[feature])):
        test_dataset[feature].replace(
            categorical_features_unique_values[feature][index][1]
            , categorical_features_unique_values[feature][index][0]
            , inplace=True
        )

test_dataset.dtypes
````

````python
# Filling Missing Values with Zeros
test_dataset.isnull().sum().sort_values(ascending=False)

features_with_missing_values = [
    'PoolQC' , 'MiscFeature', 'Alley'
    , 'Fence', 'FireplaceQu', 'LotFrontage'
    , 'GarageYrBlt', 'GarageCond', 'GarageType'
    , 'GarageFinish', 'BsmtExposure', 'BsmtFinType2'
    , 'BsmtCond', 'BsmtFinType1', 'BsmtQual'
]

test_dataset.fillna(0, inplace=True)
````
